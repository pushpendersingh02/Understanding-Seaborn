# BAR PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
sns.barplot(x=df.who,y=df.fare)
# Another way to plot graph in seaborn, here no need to give data parameter seperately.
```

```python
sns.barplot(x='class',y='fare',data=df)
plt.show()
# Using show function of matplotlib to remove Axis.
```

```python
sns.barplot(x='class',y='fare',data=df,hue='sex')
plt.show()
# Using show function of matplotlib to remove Axis.
```

```python
sns.barplot(x='class',y='fare',data=df,hue='sex')
plt.show()
```
